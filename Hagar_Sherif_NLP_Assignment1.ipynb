# Check for the quality of stemming of NLTK vs lemmatization of spacy 

**1. Stemming of NLTK**





```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import pandas as pd 

ps = PorterStemmer()

text = """Nikola Tesla was a Serbian-American inventor, electrical engineer, mechanical engineer, and futurist best known for his contributions to the design of the modern alternating current electricity supply system.""" 
tokens = word_tokenize(text)

stemmed = []

for item in tokens:
     stemmed_word = ps.stem(item)
     stemmed.append(stemmed_word)

df = pd.DataFrame(data={"tokens":tokens, "stemmed": stemmed})  
df

```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
  
group = ["Read", "Reading", "Reader", "Readers","Friends", "Friendship", "Friendly"]
  
for i in group:
    print(i, " --> ", ps.stem(i))
```

**2. Lemmatization of spacy**




```python
import spacy

nlp = spacy.load("en_core_web_sm")
text = nlp("Nikola Tesla was a Serbian-American inventor, electrical engineer, mechanical engineer, and futurist best known for his contributions to the design of the modern alternating current electricity supply system." )
list1 = []
for item in text:
  list1.append(item.lemma_)

final_text = ' '.join(map(str,list1))
print(final_text)
```

**Conclusion:**

*  Stemming reduces the word to its stem by removing all affixes.
*  Lemmatization reduces the word to its root.
